```python
#converting dataset into a data frame using pandas
cancer_df = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```python
# defining output and input
X= cancer_df.drop('target', axis=1)
y = cancer_df['target']
```

```python
X
```

```python
X.shape
```

```python
print(y, end= ' ')
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size=0.20, random_state= 5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#model training 
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC(C=1.0, kernel='rbf', degree=3, gamma='auto', coef0=0.0, shrinking=True, probability=False, tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=- 1, decision_function_shape='ovr', break_ties=False, random_state=None)
svc_model.fit(X_train, y_train)
```

```python
#result
y_predict= svc_model.predict(X_test)
cm =confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot= True)
```

```python
print(classification_report(y_test, y_predict))
```

```python
# Improving model
min_X = X_train.min()
min_X
```

```python
range_X = (X_train - min_X).max()
range_X
```

```python
X_train_scaled= (X_train - min_X)/range_X
X_train_scaled
```

```python
min_X = X_test.min()
min_X
```

```python
range_test =(X_test - min_X).max()
```

```python
X_test_scaled=(X_test - min_X)/range_test
X_test_scaled
```

```python
svc_model = SVC(C=1.0, kernel='rbf', degree=3, gamma='auto', coef0=0.0, shrinking=True, probability=False, tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=- 1, decision_function_shape='ovr', break_ties=False, random_state=None)
svc_model.fit(X_train_scaled, y_train)
```

```python
y_predict= svc_model.predict(X_test_scaled)
cm =confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot=True)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']} 
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_prediction= grid.predict(X_test_scaled)
```

```python
cm= confusion_matrix(y_test,grid_prediction)
```

```python
sns.heatmap(cm, annot= True, fmt= 'd')
```

```python
print(classification_report (y_test, grid_prediction))
```
